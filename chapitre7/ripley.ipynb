```python
import pandas as pd
import numpy as np
import pointpats
import matplotlib.pyplot as plt
```

```python
bocage = pd.read_csv('bocage_ign_2012.csv')
openfield = pd.read_csv('openfield_ign_2012.csv')
```

```python
bocage.info()
```

```python
bpp = pointpats.PointPattern(bocage.iloc[:,:])
brealizations = pointpats.PoissonPointProcess(bpp.window, bpp.n, 100, asPP=True)
bLenv = pointpats.distance_statistics.Lenv(bpp,  realizations=brealizations, intervals=50, dmin=0, dmax=15)
```

```python
bLenv.plot()
plt.ylim((-5,25))
plt.savefig('figure2_7_bocage_petit_rayon.pdf', format='pdf')
```

```python
opp = pointpats.PointPattern(openfield.iloc[:,:])
orealizations = pointpats.PoissonPointProcess(opp.window, opp.n, 100, asPP=True)
oLenv = pointpats.distance_statistics.Lenv(opp,  realizations=orealizations, intervals=50, dmin=0, dmax=15)
```

```python
oLenv.plot()
plt.ylim((-5,25))
plt.savefig('figure2_7_openfield_petit_rayon.pdf', format='pdf')
```

```python
oLenv_g = pointpats.distance_statistics.Lenv(opp,  realizations=orealizations, intervals=10, dmin=0, dmax=2000)
```

```python
oLenv_g.plot()
plt.ylim((-5,25))
plt.savefig('figure2_7_openfield_grand_rayon.pdf', format='pdf')
```
